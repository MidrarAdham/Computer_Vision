### Operators, Mechanics/Technicians, and Engineers
* Imagine that you're driving a car and it breaks down
* Do you have any idea how to determine what is wrong and whether you can fix it?
    * An operator (driver) could not
    * A mechanic/technician might
    * An automotive engineer probably could
* If there is a design problem, could you fix the design of the car?
    * An engineer could
    * The others couldn't
* These days there are many resources that enable one to get started with CV
    * OpenCV

### Conceptual Distinctions
* Technicians operate
* Engineers know how it operates
* Why choose engineering
    * You understand how it works
    * Skilled operation
    * You can design
    * You can debug
    * You know the key ideas
* Example: image rotation
    * Operator: 
        * Result looks good
        * Result doesn't look good
            * Start asking Google/ChatGPT/Stack Overflow
            * May require hours, and still not get to a solution
    * Engineer:
        * How was the image resampled?
        * How were edge conditions handled?
        * How was the image rescaled?
        * How was the resolution chosen?
        * How was color represented for processing?

```python
import cv2
import os
import matplotlib.pyplot as plt

image_path = os.path.join(os.getcwd(),'Bourbon.jpeg')
image_BGR = cv2.imread(image_path)
image_RGB = cv2.cvtColor(image_BGR, cv2.COLOR_BGR2RGB)

# Rotate the image_RGB by 20 degrees
num_rows, num_cols = image_RGB.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), -20, 1)
image_RGB_rotation = cv2.warpAffine(image_RGB, rotation_matrix, (num_cols, num_rows))

# Show the original image_RGB and image_RGB_rotation side by side in the same figure
fig = plt.figure(figsize=(16, 6))
ax1 = fig.add_subplot(1, 2, 1)
ax1.imshow(image_RGB)
ax1.set_title('Original Image')
ax2 = fig.add_subplot(1, 2, 2)
ax2.imshow(image_RGB_rotation)
ax2.set_title('Rotated Image')

plt.show() # Prevents text from being printed in the output cell before the figure
```

### Computer Vision Scope
* Computer vision has immense scope

* If I teach deep, we won't get very far

    * If an algorithm doesn't work, you won't know what to do beyond "guess and check" or plead for help with Google/Bard/ChatGPT
* In most courses, I can provide a thorough mathematical foundation
